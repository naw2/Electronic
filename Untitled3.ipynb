```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For inline plotting in Jupyter
%matplotlib inline
```

```python
# Set random seed for consistency
np.random.seed(42)

# Create dates for a month
dates = pd.date_range(start='2025-06-01', periods=30)

# Simulate random sales data (daily total sales in $)
sales = np.random.randint(100, 1000, size=30)

# Simulate sales by category (3 categories)
categories = np.random.choice(['Electronics', 'Clothing', 'Food'], size=30)

# Combine into DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Sales': sales,
    'Category': categories
})

df.head()
```

```python
# Summary statistics
print("Summary Statistics:")
print(df.describe())

# Total sales
print("\nTotal Sales:", df['Sales'].sum())

# Sales by category
print("\nSales by Category:")
print(df.groupby('Category')['Sales'].sum())
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Sales'], marker='o')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=df['Category'], y=df['Sales'], estimator=sum)
plt.title('Total Sales by Category')
plt.ylabel('Total Sales ($)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Sales'], bins=10, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Sales Spread by Category')
plt.show()
```
